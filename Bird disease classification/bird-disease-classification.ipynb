```python
from lobe import ImageModel

model = ImageModel.load(r'../input/bird-classification-using-resnet50-v2/bird TensorFlow 2')
```

```python
import os
import shutil

src = '../input/bird-species-classification/train_data/train_data/'
dest = './'

for path, subdirs, files in os.walk(src):
    for name in files:
        filename = os.path.join(path, name)
        shutil.copy2(filename, dest)
```

```python
import shutil, random, os
dirpath = './'
destDirectory = './test'

filenames = random.sample(os.listdir(dirpath), 9)
for fname in filenames:
    srcpath = os.path.join(dirpath, fname)
    shutil.copyfile(srcpath, destDirectory)
filenames
```

```python
from PIL import Image
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
for a in filenames:
    img = cv2.imread('./{}'.format(a))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    result = model.predict(Image.fromarray(img, 'RGB'))
    # Print top prediction
    print('Model Prediction : {}'.format(result.prediction))
    # Print all classes
    for label, confidence in result.labels:
        print(f"{label}: {round(confidence*100,2)}%")
```
